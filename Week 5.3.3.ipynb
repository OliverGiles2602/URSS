```python
## Attemping a sin curve 

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Defining my sin function 
# I made sure to add the possible multiples as I thought it would be interesting to mess with later

def sino(x,A,B):
    return  A*(np.sin(B*x))
```

```python

#linspace fills between the minimum and maximum with n number of points
## I increase all of the linspace variables otherwise the function would not have been seen at all

x = np.linspace(-10,10, 1000)

# We use all the points from the linspace and plot them on the sin curve of choice

y = sino(x,1,1)

# Create an array of length x with numbers in the range 0 to 1. This is our noise term.

noise = np.random.normal(0, 1, len(x))

# Add noise to our "pure" signal

data = y + noise

# Plotting
fig = plt.figure(figsize=(14.0,10.0))
plt.scatter(x, data, label='Data (with noise)', color = "royalblue")
plt.legend(loc="upper left")
```

```python
# Now using scipy to find the best way to plot this

# We use the scipy optimize package to perform our curve fit
popt, pcov = curve_fit(sino, x, data)

# popt - The best fit parameters
# pcov - The covariance matrix

print(popt)
print(" ")
print(pcov)
```

```python

fig = plt.figure(figsize=(14.0,10.0))
plt.scatter(x, data, label='data')
plt.plot(x, sino(x, *popt), 'r--', label="Fit")
plt.plot(x, y, 'lime', label="True line")
plt.legend(loc="upper left")
```

This was my favourite one to do, I spent some time trying different combinations to get interesting results.
I set out looking for a way to maximise/ see how good scipy actually was about finding the best fitted line.

```python
#linspace fills between the minimum and maximum with n number of points
## I increase all of the linspace variables otherwise the function would not have been seen at all

x = np.linspace(-10,10, 1000)



# We use all the points from the linspace and plot them on the sin curve of choice
## The variables that I change to get interesting results 
y = sino(x,3,5)

# Create an array of length x with numbers in the range 0 to 1. This is our noise term.

noise = np.random.normal(0, 1, len(x))

# Add noise to our "pure" signal

data = y + noise

popt, pcov = curve_fit(sino, x, data)
print(popt)


fig = plt.figure(figsize=(14.0,10.0))
plt.scatter(x, data, label='data',color="darkgreen")
plt.plot(x, sino(x, *popt), 'r--', label="Fit")
plt.plot(x, y, 'lime', label="True line")
plt.legend(loc="upper left")
```

Here you can see it still plots a sine curve but it oscillates so quickly that it is nowhere near correct.
I then decided to see what happens in the opposite case of a very slow sine curve.

```python
#linspace fills between the minimum and maximum with n number of points
## I increase all of the linspace variables otherwise the function would not have been seen at all

x = np.linspace(-100,100, 750)



# We use all the points from the linspace and plot them on the sin curve of choice
## The variables that I change to get interesting results 
y = sino(x,1,0.25)

# Create an array of length x with numbers in the range 0 to 1. This is our noise term.

noise = np.random.normal(-0.1, 0.3, len(x))

# Add noise to our "pure" signal

data = y + noise


popt, pcov = curve_fit(sino, x, data)
print(popt)


fig = plt.figure(figsize=(14.0,10.0))
plt.scatter(x, data, label='data',color="orange")
plt.plot(x, sino(x, *popt), 'r--', label="Fit")
plt.plot(x, y, 'purple', label="True line")
plt.legend(loc="upper left")
```
